## Consigna
Vas a crear un programa que le pida al usuario que ingrese un texto de al menos 10 palabras. Tu programa va a procesar ese texto, lo va a procesar, y le va a devolver un análisis detallado, que incluya:
1. Contar el número total de caracteres en el texto
2. Contar el número de caracteres sin incluir los espacios
3. Contar la cantidad de vocales que hay en el texto
4. Contar el número total de palabras en el texto ingresado

**nota**: para resolver los siguientes items, deberás investigar otros métodos de strings que no hemos visto. Recuerda valerte de *type*, *dir* y *help*

5. Eliminar la primera palabra
6. Reemplazar todos los espacios por guiones medios (-)
8. Cambia las mayúsculas a minúsculas y las minúsculas a mayúsculas

Al finalizar el procesamiento, se imprimirán 8 cadenas literales mostrando los resultados del análisis (una cadena por cada resultado).

### Resolución

Ingreso del usuario:

```python
texto_procesar = input("Hola, escribe un texto de más de diez palabras: ")
```

1. Contar el número total de caracteres en el texto

```python
numero_caracteres = len(texto_procesar)
print(f"El número de caracteres de tu frase es :{numero_caracteres}")
```

2. Contar el número de caracteres sin incluir los espacios

```python
texto_sin_espacios = texto_procesar.replace(" ","")
caracteres_sin_espacios = len(texto_sin_espacios)
print(f"el número de carácteres de tu frase sin espacios es: {caracteres_sin_espacios}")
```

3. Contar la cantidad de vocales que hay en el texto

```python
numero_vocales= texto_procesar.count("a") + texto_procesar.count("e") + texto_procesar.count("i") + texto_procesar.count("o") + texto_procesar.count("u")
print(f"el número de vocales de tu frase es: {numero_vocales}")
```

4. Contar el número total de palabras en el texto ingresado

```python
numero_palabras = len(texto_procesar.split())
print(f"el número de palabras de tu frase es: {numero_palabras}")
```

5. Eliminar la primera palabra

```python
lista_palabras = texto_procesar.split()
sin_primera_palabra_str = " ".join(lista_palabras[1:])
print(f"El texto sin la primera palabra es: {sin_primera_palabra_str}")
```

6. Reemplazar todos los espacios por guiones medios (-)

```python
texto_separados_guiones = texto_procesar.replace(" ","-")
print(f"El texto con guiones es: {texto_separados_guiones}")
```

7. Cambia las mayúsculas a minúsculas y las minúsculas a mayúsculas

```python
texto_letras_invertidas = texto_procesar.swapcase()
print(f"El texto con letras invertidas es: {texto_letras_invertidas}")
```
